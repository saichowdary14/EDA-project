## Problem statement

The project aims to analyze the used bike market by examining key factors that influence the pricing, popularity, and condition of bikes available for sale. With increasing demand for used bikes in various cities, there’s a need to understand the trends and behaviors in the used bike market to help both buyers and sellers make informed decisions. The goal is to provide insights on how various features (such as kilometers driven, brand, year, and location) impact the pricing of bikes, and to identify patterns that can guide better pricing strategies and demand forecasting in real-time

```python
year=[]
Brand_name=[]
Kilometers=[]
price=[]
used=[]
Petrol=[]
url="https://www.bikedekho.com/used-bikes+in+hyderabad"
driver.get(url)

last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3) 

    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break 
    last_height = new_height
html=driver.page_source
soup=BeautifulSoup(html)
data=soup.find_all("div",class_="gsc_col-sm-12 gsc_col-md-12 trustmarkbox holder")

for i in data:
    regex=re.findall(r"\b\d{4}\b",i.text)
    if len(regex)>0:
        year.append(regex[0])
    else:
        year.append(np.nan)
for i in data:
    regex=re.findall(r"(?<=\d{4}\s)[A-Za-z]+\s[A-Za-z]+\s?[a-zA-Z0-9]+\b",i.text)
    if len(regex)>0:
        Brand_name.append(regex[0])
    else:
        Brand_name.append(np.nan)
for i in data:
    regex=re.findall(r"\d+[,]\d+\s[k][m]",i.text)
    if len(regex)>0:
        Kilometers.append(regex[0])
    else:
        Kilometers.append(np.nan)
for i in data:
    regex=re.findall(r"[R][s]\.[0-9\.,]+\s?[L]?",i.text)
    if len(regex)>0:
        price.append(regex[0])
    else:
        price.append(np.nan)
for i in data:
    regex=re.findall(r"(First|Second)",i.text)
    if len(regex)>0:
        used.append(regex[0])
    else:
        used.append(np.nan)
for i in data:
    regex=re.findall(r"(Petrol|Electric)",i.text)
    if len(regex)>0:
        Petrol.append(regex[0])
    else:
        Petrol.append(np.nan)
    
    
```

**finding outliers**

```python
q1=df["Price"].quantile(0.25)
q3=df["Price"].quantile(0.75)
iqr=q3-q1
lower_bound=q1-(1.5*iqr)
upper_bound=q3+(1.5*iqr)
lower_outliers=[]
upper_outliers=[]
for val in df["Price"]:
    if val<lower_bound:
        lower_outliers.append(val)
    elif val>upper_bound:
        upper_outliers.append(val)
print(f"lower outliers:{lower_outliers}")
print(f"upper outliers:{upper_outliers}")
    
```

**Replacing the outliers by CAPPING method**

```python
q1=df["Price"].quantile(0.25)
q3=df["Price"].quantile(0.75)
iqr=q3-q1
lower_bound=q1-(1.5*iqr)
upper_bound=q3+(1.5*iqr)
lower_replace=df["Price"].quantile(0.05)
upper_replace=df["Price"].quantile(0.95)
df["Price"] = df["Price"].apply(lambda x: lower_replace if x < lower_bound else (upper_replace if x > upper_bound else x))
```

## Meaningful Insights:
### Kilometers vs Price:

As the number of kilometers a bike has driven increases, its price tends to decrease.
This helps buyers know that bikes with fewer kilometers will be more expensive.
### Bike Age:

Newer bikes (less than 5 years old) tend to have fewer kilometers and higher prices.
Buyers looking for newer bikes should expect to pay more.
### Location Impact:

Most bikes are sold in Bangalore and Hyderabad, and these cities have higher bike prices.
Buyers and sellers in these cities can expect more demand and higher prices.
### Brand Influence:

Royal Enfield bikes have higher prices, while Bajaj is also a popular brand.
Buyers can expect to pay more for Royal Enfield bikes due to their higher demand.
### Fuel Type:

Petrol bikes generally cost more than electric bikes.
This shows that petrol bikes are still more popular in the used market.
### Ownership:

Most bikes are owned by first owners (93%).
First-owner bikes are often considered in better condition and are more sought after.

## Conclusion:
This analysis helps understand how different factors—such as kilometers, bike age, brand, and location—affect the price of used bikes. The main takeaway is that newer bikes with fewer kilometers are priced higher, and certain brands like Royal Enfield are more expensive. Buyers can use this information to make smarter decisions, while sellers can price their bikes based on the bike's condition and location.

In simple terms, understanding these patterns can help both buyers and sellers navigate the used bike market more effectively.
